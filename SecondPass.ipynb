```python
helloFile = open('LUNAR_LANDING_GUIDANCE_EQUATIONS.agc.txt')
```

```python
helloContent = helloFile.read()
```

```python
lines=helloContent.split('\n')
```

```python
len(lines)
```

```python
lines
```

```python
import re
```

```python
comments = []
for line in lines:
    if line.strip(): #delete blank lines
        comments.append(line)
    else:
        #print(line) removing all blank lines
        pass
    
```

```python
comments
```

```python
len(comments)
```

```python
phrases=[]
for comment in comments:
    x = re.split("#", comment)
    for phrase in x:
        if len(phrase.strip())!=0:
            #replace consecutive tabs or spaces with a single
            phrase = re.sub("[\t]{1,}", "   ", phrase.strip()) 
            #add to list
            phrases.append(phrase)
        
```

```python
phrases
```

```python
len(phrases)
```

```python
#save out all phrases used
import csv
with open('sortedphrases.csv', 'w') as f:
    for phrase in phrases:
        f.write("%s\n"%(phrase))
```
